```python
from PIL import Image  
import matplotlib.pyplot as plt  

def median_filter(image, mask_size=3):
    width, height = image.size
    filtered_image = Image.new("L", (width, height))  
    padding = mask_size // 2  

    for x in range(padding, width - padding):
        for y in range(padding, height - padding):
            pixels = []
            for i in range(-padding, padding + 1):
                for j in range(-padding, padding + 1):
                    pixel = image.getpixel((x + i, y + j))
                    pixels.append(pixel)

            
            pixels.sort()
            median = pixels[len(pixels) // 2]

            filtered_image.putpixel((x, y), median)

    return filtered_image

def compute_difference(image1, image2):
    width, height = image1.size
    diff_image = Image.new("L", (width, height))  

    for x in range(width):
        for y in range(height):
            pixel1 = image1.getpixel((x, y))
            pixel2 = image2.getpixel((x, y))
            diff = abs(pixel1 - pixel2)  
            diff_image.putpixel((x, y), diff)

    return diff_image

input_image = Image.open("image1.png").convert("L")

filtered_image = median_filter(input_image)

diff_image = compute_difference(input_image, filtered_image)

filtered_image.save('filtered_image.png')
diff_image.save('diff_image.png')

plt.figure(figsize=(10, 8))

plt.subplot(1, 3, 1)
plt.title("Исходное изображение")
plt.imshow(input_image, cmap='gray')
plt.axis('off') 
plt.subplot(1, 3, 2)
plt.title("Отфильтрованное изображение")
plt.imshow(filtered_image, cmap='gray')
plt.axis('off') 
plt.subplot(1, 3, 3)
plt.title("Разностное изображение")
plt.imshow(diff_image, cmap='gray')
plt.axis('off') 
plt.show()
```
